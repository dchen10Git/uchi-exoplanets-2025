```python
import rebound
import time

# Create simulation
sim = rebound.Simulation()
sim.add(m=1.0) # star
sim.add(m=1e-3, a=1.0, e=0.2)
sim.add(m=1e-3, a=0.5, e=0.3)
sim.add(m=1e-3, a=1.5, e=0)

sim.move_to_com()

# Launch widget
sim.widget(size=(400,400))
```

```python
for i in range(5000):
    sim.integrate(sim.t + 0.02)   # advance time a little
    time.sleep(0.01)
```
